```python
if dfp.iloc[-1][0]=='Grand Total:':
    dfp = dfp[:-1].reset_index(drop = True)
else:
    pass

dfp = dfp[13:].reset_index(drop=True)
# Change column names to the first row values
dfp.columns = dfp.iloc[0]
dfp = dfp[1:].reset_index(drop=True)
```

#### New Stuff - make sure to copy in boilerplate!

```python
dfp['Date']=pd.to_datetime(dfp['Date'])
#dfp.dtypes

### IN CASE DATE IS IN mm/dd/yyyy, following code shows how to make a proper comparison
### if you just compare the date column with '2020-02-05', the result will return NaN because Date format is not consistent
limit_time = pd.to_datetime('2/5/2020')
dfp = dfp.loc[dfp['Date']>limit_time,:].sort_values(by='Date').reset_index(drop=True)
```

```python
## assign language
dfp.loc[dfp['Placement'].str.contains('FR'),'Language']='FR'
dfp['Language'].fillna('EN',inplace=True)

## assign page
dfp.loc[dfp['Placement'].str.contains('Homepage'),'Page']='Homepage'
dfp.loc[dfp['Placement'].str.contains('homepage'),'Page']='Homepage'
dfp.loc[dfp['Placement'].str.contains('Detail'),'Page']='VDP'
dfp.loc[dfp['Placement'].str.contains('detail'),'Page']='VDP'

dfp['ID']=dfp['Language'].map(str) + dfp['Page'].map(str)

dfp=dfp[dfp['Page'].notnull()].reset_index(drop=True)
dfp.shape
```
